#Challenge 1: Load a CSV file into a Pandas Dataframe.

```python
#lets load our data set
df = pd.read_csv('Nigeria economy kpis.csv')
df.head()
```

#Challenge 2: Read a dataframe metadata/general information.


```python
# Read a dataframe metadata/general information.
df.info()
```

#Challenge 3: Perform descriptive statistic analysis on numeric data.


```python
#Perform descriptive statistic analysis on numeric data.
df.describe()
```

# Challenge 4: Rename a dataset columns.


```python
df.rename(columns= {'Year':'year', 'Inflation rate': 'Inflation_rate', 'Unemployment': 'Unemployment rate', 'Government debt': 'Government debt rate', 'Agriculture': 'Agriculture GDP (Naira)', 'Industry': 'Industry GDP (Naira)', 'Services': 'Services GDP (Naira)', 'Leadership': 'Leadership'}, inplace=True)
#display the fisrt 2 rows
df.head(2)

```

# Challenge 5: Apply transformations on multiple columns.


```python
#transform the inflation rate records into percentage
df['Inflation_rate'] = df['Inflation_rate'].apply(lambda x: x*100)
df.head()
```

```python
#transform the 'Government debt rate 'and 'Unemployment rate' records into percentage
df[['Unemployment rate', 'Government debt rate']] = df[['Unemployment rate', 'Government debt rate']].apply(lambda x: x*100)
df.head()
```

#Challenge 6: Apply basic filters.

```python

# select all records where 'Olusegun Obasanjo' is the president of the country
df[df['Leadership'] == 'Olusegun Obasanjo']
```

# Challenge 7: Apply advanced filters.


```python
#in which year did the inflation_rate record its highest value
df.loc[df['Inflation_rate']==df['Inflation_rate'].max()]
```

# Challenge 8: Group the data with a specific key.


```python
# lets check the average unemployment rate for each precident era
df.groupby('Leadership')[['Unemployment rate','Inflation_rate','Government debt rate']].mean().plot(kind='bar').set_title('Average Unemployment rate by President')
```

#Challenge 9: Build simple visualizations to analyse the output.


```python
#display the last output using a bar chart
df.groupby('Leadership')['Unemployment rate'].mean().plot(kind='bar', title='Average Unemployment rate by President')

```

```python
#lets create a visualization of  the uneployment rate over the years
df.groupby('year')['Unemployment rate'].mean().plot(kind='line', title='Unemployment rate over the years')

```
